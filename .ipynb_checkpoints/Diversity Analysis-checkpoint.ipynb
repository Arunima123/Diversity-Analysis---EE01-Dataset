```python
#https://www.eeoc.gov/statistics/employment/jobpatterns/eeo1

import pandas as pd
import csv
      
def getColumnMappingDict():
    input_file = csv.DictReader(open("ColumnHeaderInformation.csv"))
    column_mapping = {}
    for row in input_file:
        column_label_split = row['Column Label'].split(",")
        column_label = {}
        for label in column_label_split:
            label_split = label.split(":")
            column_label[label_split[0]]=label_split[1]
        column_mapping[row['Column Name']]=column_label
    return column_mapping

def getColumnHeadersWithAggregates(column_mapping):
    column_headers_agg = []
    for key,value in column_mapping.items():
        print(value['Race'], (value['Race']=='Aggregate'))
        if(value['Race']=='Aggregate'):
            column_headers_agg.append(key)
    return column_headers_agg
```

```python
column_mapping = getColumnMappingDict()
column_headers_agg = getColumnHeadersWithAggregates(column_mapping)
```
