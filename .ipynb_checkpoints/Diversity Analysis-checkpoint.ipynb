```python
#https://www.eeoc.gov/statistics/employment/jobpatterns/eeo1

import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import table

column_mapping_file = "ColumnHeaderInformation.csv"
EE01_dataset = "DiversityDataset.csv"
column_aggregate_all = 'TOTAL10'
industry_name = "NAICS2_Name"

def getColumnMappingDict():
    input_file = csv.DictReader(open(column_mapping_file))
    column_mapping = {}
    for row in input_file:
        column_label_split = row['Column Label'].split(",")
        column_label = {}
        for label in column_label_split:
            label_split = label.split(":")
            column_label[label_split[0].strip()]=label_split[1].strip()
        column_mapping[row['Column Name'].strip()]=column_label
    return column_mapping


def getColumnHeadersWithParticularLabel(column_mapping, label):
    column_headers_agg_race = []
    column_headers_agg_sex = []
    column_headers_agg_job = []
    for key,value in column_mapping.items():
        if value['Race']==label:
            column_headers_agg_race.append(key)
        elif value['Sex']==label:
            column_headers_agg_sex.append(key)
        elif value['Job']==label:
            column_headers_agg_job.append(key)
    return column_headers_agg_race, column_headers_agg_sex, column_headers_agg_job

def getDataFrameFromCSV(file_name):
    df = pd.read_csv(file_name, encoding = "ISO-8859-1", engine='python')
    df.columns = df.columns.str.strip()
    return df

def getRelevantColumns(df, column_list):
    print("Using Columns from EE01 Dataset - ", column_list)
    df = df[column_list]
    return df

def plotPieChartForADataFrame(df, title, figsizeX, figsizeY, yLabel, xLabel):
    plot = df.plot.pie(title=title, 
                   subplots=True, 
                   figsize=(figsizeX, figsizeY), 
                   legend=None, 
                   ylabel=yLabel,
                   xlabel=xLabel,   
                  autopct='%1.1f%%')

```

```python
column_mapping = getColumnMappingDict()
column_headers_agg_race, column_headers_agg_sex, column_headers_agg_job = getColumnHeadersWithParticularLabel(column_mapping, 'Aggregate')
EE01_df = getDataFrameFromCSV(EE01_dataset)
```

```python
# Step 1 : State wise - aggregate comparision
relevant_columns = [industry_name, column_aggregate_all]
df = getRelevantColumns(EE01_df, relevant_columns)
df = df.groupby(industry_name)
df = df.aggregate(np.sum)
```
